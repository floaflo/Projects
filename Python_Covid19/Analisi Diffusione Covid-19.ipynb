2. chiede di trovare, per ogni continente:
 a. il numero di casi fin dall'inizio della pandemia
 b. la percentuale rispetto al totale mondiale del numero di casi

```python
#esercizio 2.a: per trovare il numero totale di casi uso group by per continente 
#in una nuova variabile che chiamo casi_per_continente. Uso la colonna new cases per evitare il double couting come nella build week di Excel

casi_per_continente = covid.groupby ("continent")["new_cases"].sum().sort_values (ascending = False)

print ("Questo è il numero totale di casi per continente, dal continente con più casi a quello con meno casi:", "\n", "\n", casi_per_continente)

print ("\n", "I numeri per continente mostrano che l'Africa risulta il continete con il minor numero di casi, questo può essere spiegato anche dal fatto che il tracciamento dei casi in questo continente è stato meno regolare")
```

```python
#esercizio 2.b: la percentuale rispetto al totale mondiale del numero di casi
#parto dal capire il totale mondiale che userò per la divisione, sommando tutti i numeri di prima

totale_mondiale = casi_per_continente.sum()
print ( "Questo è il numero totale di casi:", totale_mondiale) 

percentuale_per_continente = (casi_per_continente / totale_mondiale) *100

for continent,percent in percentuale_per_continente.items():
    print(f"{continent}: {round(percent,2)} %") #uso un ciclo for per stamparmi le % e arrotondo
```

3. Selezionare i dati relativi all'Italia nel 2022 e, poiché i nuovi casi vengono registrati settimanalmente, filtrare via i giorni che non hanno misurazioni; quindi mostrare con dei grafici adeguati:
 a. l 'evoluzione del casi totali dall'inizio alla fine dell'anno
 b. il numero di nuovi casi rispetto alla data

```python
#come primo passo mi importo seaborn e matplot per i grafici

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#esercizio 3.a: l 'evoluzione del casi totali dall'inizio alla fine dell'anno e grafico

#creo il filtro paese Italia e anno 2022 
#prima converto date in formato data perchè è un object e il filtro mi da errore

covid["date"] = pd.to_datetime(covid["date"])

casi_italia_2022 = covid[(covid["location"] == "Italy")&(covid["date"].dt.year == 2022)]

#rimuovo dove non ci sono state misurazioni in new cases
casi_italia_2022 = casi_italia_2022[casi_italia_2022["new_cases"] > 0]

#nell'asse y voglio il numero di casi covid, nell'asse x voglio i mesi del 2022. 
#metto le date come asse x e rendo il grafico carino
casi_italia_2022.set_index("date")["total_cases"].plot(marker ="o", figsize=(15, 7))
plt.title("Grafico 1: Andamento casi covid nel 2022 in Italia", fontsize = 16) 
plt.ylabel("Numero casi", fontsize= 13)
plt.xlabel ("Mesi anno 2022", fontsize= 13)
plt.ticklabel_format(style="plain", axis="y")
plt.grid(True)
plt.show()
```

```python
#esercizio 3.b: il numero di nuovi casi rispetto alla data
#ora uso lo stesso codice ma invece dei casi totali uso solo i nuovi

casi_italia_2022.set_index("date")["new_cases"].plot(marker ="o", figsize=(15, 7)) #metto la data come asse x, per farlo più carino ci metto dei pallini
plt.title("Grafico 2: Andamento nuovi casi covid nel 2022 in Italia", fontsize = 16) 
plt.ylabel("Numero casi", fontsize= 13)
plt.xlabel ("Mesi anno 2022", fontsize= 13)
plt.ticklabel_format(style="plain", axis="y") #aggiunto visto che i numeri sull asse y non erano chiari
plt.grid(True)  #aggiunta griglia per farlo più carino
plt.show()
```

Il grafico 1 mostra l'andamento totale dei casi di covid in Italia nel 2022: il numero di casi è incrementato costantemente nel corso del 2022. Il grafico 2 invece mostra in quali mesi ci sono stati i picchi maggiori di nuovi contagi, in particolare i contagi sono stati molto elevanti ad inizio 2022 per poi andare piano piano a diminuire durante il corso dell'anno, con una risalita ad inizio estate tra Luglio e Agosto. 

4. Riguardo le nazioni di Italia, Germania e Francia:
   a. mostrare in un boxplot la differenza tra queste nazioni riguardo il numero di pazienti in terapia intensiva (Intensive Care Unit, ICU, considerare quindi la colonna icu_patients) da maggio 2022 (incluso) ad aprile 2023 (incluso)
   b. scrivere un breve commento (una o due righe) riguardo che conclusioni possiamo trarre osservando il grafico risultante

```python
#esercizio 4.a

#creo una lista di paesi

paesi= ["Italy", "Germany", "France"]

icu_data = covid[
    covid["location"].isin(paesi) & #filtro solo i paesi nella lista
    (covid["date"] >= "2022-05-01") & #filtro per la data di partenza
    (covid["date"] <= "2023-04-30") & #filtro per la data di fine
    (covid["icu_patients"].notna())].copy() #elimino nulli e creo copia


sns.boxplot (data=icu_data, x= "location", y= "icu_patients", color="lightblue")
plt.title("Grafico 3: Numero pazienti terapia intensiva Maggio 2022 - Aprile 2023 per paese", fontsize=15)
plt.ylabel ("Numero pazienti",fontsize=13)
plt.xlabel ("Paesi",fontsize=13)
plt.grid(True)
plt.show()
```

esercizio 4.b

Il grafico 3 ci aiuta a confrontare i dati relativi al numero di pazienti in terapia intensiva tra Francia, Germania e Italia. Il grafico mostra che l'Italia ha avuto un numero più basso di ricoveri intesivi rispetto agli altri due paesi, infatti se si guardano i baffi (le linee che escono fuori dalla scatola blu) mentre la Germania ha avuto più di 1750 pazienti e la Francia più di 1500, l'Italia è rimasta sotto i 500. La linea centrale all interno delle scatole blu (le mediane) ci indicano che in tutti è 3 i paesi il numero di pazienti in terapia intesiva è stato equilibrato durante il pariodo osservato, ma in Italia è stato nettamente più basso rispetto a Francia e Germania.


5. Riguardo le nazioni di Italia, Germania, Francia e Spagna in tutto il 2021:
   a. mostrare, in maniera grafica oppure numerica, la somma dei pazienti ospitalizzati per ognuna colonna (hosp_patients)
   b. se ci sono dati nulli, con un breve commento scrivere se può essere possibile gestirli tramite sostituzione o meno

```python
#esercizio 5.a

paesi2= ["Italy", "Germany", "France", "Spain"]

pazienti_2021 = covid[
    (covid["location"].isin(paesi2)) &
    (covid["date"].dt.year == 2021) &
    (covid["hosp_patients"].notna())].copy()

pazienti_2021 = pazienti_2021.groupby("location")["hosp_patients"].sum().sort_values(ascending=False)
print("Questo è il numero di pazienti ospitalizzati per paese:", "\n", pazienti_2021)

pazienti_2021.plot(kind="bar", figsize=(10, 6), color="lightblue", title=" Grafico 4: Pazienti ospedalizzati nell'anno 2021 per paese")
plt.ylabel ("Numero pazienti",fontsize=13)
plt.xlabel ("Paesi",fontsize=13)
plt.ticklabel_format(style="plain", axis="y")
plt.tight_layout()
plt.show()
```

```python
#esercizio 5.b
#contiamo i nulli

nulli = (covid["location"].isin(paesi2)) & (covid["date"].dt.year == 2021)
nulli = covid[nulli]["hosp_patients"].isnull().groupby(covid["location"]).sum()
print(nulli)
```
